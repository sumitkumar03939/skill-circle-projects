# 1. Monthly Bill Tracker

```python
bills = [1500, 2000, 3000]
total = 0

for bill in bills:
    print(f"Bill: {bill}")
    total += bill

print(f"Total: {total}")
```

# 2. School Attendance

```python
students = ["John", "Mary", "Alex"]

for student in students:
    print(f"{student} Present")
```

# 3. Salary Distribution

```python
bonuses = {"John": 5000, "Mary": 6000}

for name, amount in bonuses.items():
    print(f"{name}: {amount}")
```

# 4. Grade Report

```python
scores = [85, 92, 76, 61]

for score in scores:
    if score >= 90: grade = "A"
    elif score >= 80: grade = "B"
    elif score >= 70: grade = "C"
    else: grade = "F"
    print(f"Score: {score}, Grade: {grade}")
```

# 5. Book Availability

```python
books = ["Python Basics", "Data Science"]
search = input("Enter book title: ")

found = "Not Available"
for book in books:
    if book == search:
        found = "Available"
        break
print(found)
```

# 6. Temperature Analysis

```python
temps = [30, 32, 28, 35, 29]
high, low = temps[0], temps[0]

for t in temps:
    if t > high: high = t
    if t < low: low = t

print(f"Highest: {high}, Lowest: {low}")
```

# 7. Inventory Management

```python
inventory = {"Apples": 5, "Bananas": 20}

for item, qty in inventory.items():
    if qty < 10:
        print(f"Low Stock: {item}")
```

# 8. Flight Seat Booking

```python
seats = [1, 2, 3, 4]
booked = [2, 4]

for seat in seats:
    status = "Booked" if seat in booked else "Available"
    print(f"Seat {seat}: {status}")
```

# 9. Savings Interest Calculator

```python
balances = [5000, 5200, 5300]
total_interest = 0

for bal in balances:
    total_interest += bal * 0.0005

print(f"Total Interest: {total_interest}")
```
