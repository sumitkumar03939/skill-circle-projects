# 1. Hospital Emergency Room Triage

```python
temp = float(input("Temp: "))
pain = input("Pain (none/moderate/severe): ").lower()

if temp > 102 or pain == "severe":
    print("High Priority")
elif 99 <= temp <= 102 or pain == "moderate":
    print("Medium Priority")
else:
    print("Low Priority")
```

# 2. Shopping Cart Free Delivery

```python
cart = int(input("Cart Value: "))
mem = input("Member (yes/no): ").lower()

if cart > 2000 or (1000 <= cart <= 2000 and mem == "yes"):
    print("Free Delivery")
else:
    print("Delivery Fee: ₹100")
```

# 3. Library Fine System

```python
days = int(input("Overdue Days: "))

fine = days * 2 if days <= 5 else days * 5 if days <= 10 else days * 10
print(f"Fine: ₹{fine}")
```

# 4. Airport Luggage Check

```python
cls = input("Class (economy/business/first): ").lower()
wt = float(input("Weight: "))

limit = 20 if cls == "economy" else 30 if cls == "business" else 40
rate = 200 if cls == "economy" else 150 if cls == "business" else 100

excess = max(0, wt - limit)
print(f"Excess Fee: ₹{excess * rate}")
```

# 5. Hotel Room Pricing

```python
cust = input("Type (regular/member/VIP): ").lower()
room = input("Room (standard/deluxe): ").lower()
price = 1000 # Assuming base price 1000 for calculation

if cust == "vip":
    disc = 0.30
elif cust == "regular":
    disc = 0.10
else: # Member
    disc = 0.20 if room == "deluxe" else 0.15

print(f"Final Price: {price * (1 - disc)}")
```

# 6. Energy Consumption Billing

```python
units = int(input("Units: "))

if units <= 100:
    bill = units * 5
elif units <= 300:
    bill = (100 * 5) + (units - 100) * 10
else:
    bill = (100 * 5) + (200 * 10) + (units - 300) * 15
print(f"Total Bill: ₹{bill}")
```

# 7. Ride Fare Calculation

```python
time = input("Time (day/night): ").lower()
dist = float(input("Distance (km): "))

rate = 15 if time == "day" else 20
fare = rate * dist

if dist > 20:
    fare *= 0.90 # Apply 10% discount

print(f"Total Fare: ₹{fare}")
```

# 8. Event Entry Rules

```python
age = int(input("Age: "))
student = input("Student (yes/no): ").lower()
price = 100 # Base price example

if age < 12 or age > 60:
    print("Entry: Free")
elif 12 <= age <= 25 and student == "yes":
    print(f"Entry: ₹{price * 0.5}")
else:
    print(f"Entry: ₹{price}")
```

# 9. Parking Lot Charges

```python
veh = input("Vehicle (car/bike): ").lower()
hrs = int(input("Hours: "))

if veh == "car":
    fee = 50 + max(0, hrs - 1) * 20
else:
    fee = 20 + max(0, hrs - 1) * 10
print(f"Parking Fee: ₹{fee}")
```

# 10. Real Estate Loan Eligibility

```python
inc = int(input("Income: "))
loan = input("Existing Loan (yes/no): ").lower()

if inc > 100000 and loan == "no":
    print("LTV Ratio: 90%")
elif (50000 <= inc <= 100000) or loan == "yes":
    print("LTV Ratio: 70%")
else:
    print("LTV Ratio: 50%")
```
