# 1. ATM Pin Validation

```python
correct_pin = "1234"
attempts = 0

while attempts < 3:
    pin = input("Enter PIN: ")
    if pin == correct_pin:
        print("Access Granted")
        break
    attempts += 1
else:
    print("Card Blocked")
```

# 2. Password Strength Checker

```python
while True:
    pwd = input("Password: ")
    if len(pwd) >= 8 and not pwd.isalpha() and not pwd.isdigit():
        print("Valid Password")
        break
    print("Too weak. Try again.")
```

# 3. Recharge Reminder

```python
days = int(input("Days left: "))

while days > 0:
    print(f"{days} days remaining.")
    days -= 1
print("Plan Expired")
```

# 4. Data Entry Validation

```python
while True:
    email = input("Enter Email: ")
    if "@" in email and "." in email:
        print("Valid Email")
        break
    print("Invalid. Try again.")
```

# 5. Loan Repayment Simulation

```python
loan = 5000
payment = 1000

while loan > 0:
    loan -= payment
    print(f"Balance: {max(0, loan)}")
print("Loan Repaid")
```

# 6. Quiz Game

```python
score = 0
while score < 5:
    ans = input("What is 5 + 3? ")
    if ans != "8":
        print("Game Over")
        break
    score += 1
print("You Win!")
```

# 7. Temperature Check

```python
import random
# Simulating readings
while True:
    temp = random.randint(-20, 0) 
    print(f"Reading: {temp}°C")
    if temp > -5:
        print("Warning: Threshold Breached!")
        break
```

# 8. Parking Lot Availability

```python
spaces = 10
while spaces > 0:
    input("Press Enter for Car Entry")
    spaces -= 1
    print(f"Spaces left: {spaces}")
print("Parking Full")
```

# 9. Toll Booth Counter

```python
cars = 0
while input("Enter vehicle (or 'stop'): ") != "stop":
    cars += 1
print(f"Total Vehicles: {cars}")
```

# 10. Shopping Budget Tracker

```python
budget = float(input("Budget: "))

while budget > 0:
    price = float(input("Item Price: "))
    if price > budget:
        print("Insufficient funds.")
        break
    budget -= price
    print(f"Remaining: {budget}")
```
