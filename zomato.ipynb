# Data Setup & Loading
- **Code Action:** Imported necessary libraries (pandas, numpy, matplotlib, seaborn) and
loaded the dataset Indian-Resturants.csv.

- **Data Overview:**

- The dataset contains **207,344 rows** and **26 columns.**

- Key columns include name, city, cuisines, price_range, aggregate_rating,

# Data Cleaning
- **Code Action:**

- Removed duplicate rows to ensure data integrity.

- Replaced placeholders like '-' and '-1' with NaN.

- Converted aggregate_rating to a numeric type for calculation.

- Removed rows where ratings were missing.

- **Status:** The data is now pre-processed and ready for statistical analysis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Indian-Resturants.csv", encoding="cp1252", on_bad_lines="skip")
```

```python
print(df.head())
print(df.info())
print(df.describe(include='all'))
```

```python
df = df.drop_duplicates()
df.replace(['-', '-1'], np.nan, inplace=True)
df['aggregate_rating'] = pd.to_numeric(df['aggregate_rating'], errors='coerce')
df = df[df['aggregate_rating'].notna()]
```

# Price Range vs Rating:
This analysis investigates the relationship between the cost of dining and customer satisfaction.
It aims to determine if higher prices correlate with better customer ratings.

**Graph Analysis**
The box plot below compares the distribution of aggregate_rating across four distinct
price_range categories (1 being the cheapest, 4 being the most expensive).

**Key Observations:**

- **Positive Correlation:** There is a clear upward trend in ratings as the price range increases.
The median rating (the line inside the box) for Price Range 4 is significantly higher than for
Price Range 1.

- **Quality Consistency:**

- **Price Range 1 (Budget):** Shows the widest variance in ratings (tallest box and
whiskers). While some budget places are rated highly, many are rated very poorly. This
suggests a "hit or miss" experience for customers.

- **Price Range 4 (Luxury):** Shows a tighter distribution of ratings, concentrated at the
higher end. This implies that customers can generally expect a consistent, high-quality
experience when paying a premium.

- **Outliers:** Lower price ranges have numerous outliers on the lower end, indicating a segment
of budget restaurants with very poor customer feedback.

**Visual Report**
The graph below, extracted directly from your analysis file, displays the relationship between
price and rating.

**Conclusion:** Investing in a higher price point (Premium/Luxury) appears to be a viable strategy
for ensuring higher customer satisfaction, likely due to the ability to invest more in quality
ingredients, service, and ambiance. Budget restaurants face a tougher challenge in maintaining
quality consistency.
