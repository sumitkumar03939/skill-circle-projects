# 1. Multiplication Table Generator

```python
for i in range(1, 11):
    print(*(i * j for j in range(1, 11)))
```

# 2. Prime Numbers Finder

```python
primes = [x for x in range(2, 101) if all(x % i != 0 for i in range(2, int(x**0.5) + 1))]
print(primes)
```

# 3. Seating Arrangement Checker

```python
for row in range(1, 4):
    print(f"Row {row}: " + ", ".join(f"Seat {s}" for s in range(1, 5)))
```

# 4. Unique Word Counter

```python
sentence = "hello world hello python"
unique = []
for word in sentence.split():
    if word not in unique: unique.append(word)
print(len(unique))
```

# 5. Library Book Management

```python
racks = [["Book A", "Book B"], ["Book C", "Book D"], [], [], []] # 5 racks
search = input("Search Book: ")
for i, rack in enumerate(racks):
    for book in rack:
        if book == search: print(f"Found in Rack {i+1}")
```

# 6. Pattern Printing (Triangle)

```python
for i in range(1, 6):
    print("".join(str(j) for j in range(1, i + 1)))
```

# 7. Temperature Comparison Across Cities

```python
# Data: City -> List of 7 days temps
temps = {"NY": [30, 32, 28, 25, 29, 31, 33], "LA": [40, 42, 39, 41, 43, 38, 40]}
for city, vals in temps.items():
    print(f"{city}: Max {max(vals)}, Min {min(vals)}")
```

# 8. Password Strength Checker

```python
pwds = ["pass", "Pass123!"]
for p in pwds:
    if len(p) >= 8 and any(c.isdigit() for c in p) and any(not c.isalnum() for c in p):
        print(f"{p}: Strong")
    else:
        print(f"{p}: Weak")
```

# 9. Inventory Tracker

```python
warehouse = [[10, 2, 15], [4, 20, 3]] # Sections with item counts
for i, section in enumerate(warehouse):
    for qty in section:
        if qty < 5: print(f"Low stock in Section {i+1}: {qty} units")
```

# 10. Tic-Tac-Toe Grid

```python
grid = [["_" for _ in range(3)] for _ in range(3)]
while True:
    [print(row) for row in grid]
    r, c = map(int, input("Row Col (0-2): ").split())
    grid[r][c] = "X"
```

# 11. Palindrome Checker (Strings)

```python
words = ["level", "world", "radar"]
for w in words:
    if w == w[::-1]: print(f"{w} is Palindrome")
```

# 17. Magic Square Validator

```python
sq = [[2, 7, 6], [9, 5, 1], [4, 3, 8]]
target = sum(sq[0])
is_magic = all(sum(r) == target for r in sq) and all(sum(c) == target for c in zip(*sq))
print("Magic Square" if is_magic else "Not Magic")
```
