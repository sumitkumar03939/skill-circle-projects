# 7. Leap Year Checker

```python
y = int(input("Enter year: "))
print("Leap Year" if (y % 4 == 0 and y % 100 != 0) or (y % 400 == 0) else "Not a Leap Year")
```

# 17. BMI Calculator

```python
w, h = float(input("Weight (kg): ")), float(input("Height (m): "))
bmi = w / (h ** 2)
print("Underweight" if bmi < 18.5 else "Normal" if bmi <= 24.9 else "Overweight")
```

# 18. Weekend Check

# 19. Flight Fare Check

```python
age, fare = int(input("Age: ")), float(input("Fare: "))
print(f"Fare: {fare * 0.5}" if age < 12 or age > 60 else f"Fare: {fare}")
```

# 20. Electricity Bill
